# Cleaning dataset section

- At first, as I said, I want to sellect those columns which are needed for our model from spi_matches dataset.

```python
newdataset= spi_matches.loc[:,'team1':'score2']     # It sellect any rows and column 'team1' to 'score2'
newdataset.head()
```

- Now I want to remove those rows which have some missing data( or NaN data)

```python
newdataset=newdataset.dropna(how='any')      # for removing NaN data rows
newdataset.isna()
```

- So, now after removing NaN rows the number of rows have been decreased. For some quick summary next code is provided.

```python
newdataset.describe()
```

# Merging Datasets

- Since we just have current offensive and deffensive power for teams and the spi_matches dataset is back to 2016. The SPI power score is different each year. But,suppose that offensive power is constant for each team during last 3 years. Now I want to merge 2 dataset to reach out to a new dataset. For doing this, I want to add the offensive power for team1 and team2 to the spi_matches dataset. 

```python

merged_dataset_1 = pd.merge(spi_off_def[['name','off']] ,newdataset, left_on='name' , right_on='team1' , how='left' )
merged_dataset=pd.merge(merged_dataset_1 ,spi_off_def[['name','off']] , left_on='team2' , right_on='name' , how='right' )
merged_dataset.head()
```

- Now rename the 2 columns which added recently.

```python
merged_dataset.rename(columns={'off_x':'offensive_team1'}, inplace=True)
merged_dataset.rename(columns={'off_y':'offensive_team2'}, inplace=True)
merged_dataset.columns       # shows the columns of the dataset
```

- Now I want to select those columns which are needed and change their order to make the dataset more clear.

```python
merged_dataset=merged_dataset[['team1','team2','offensive_team1','offensive_team2','spi1','spi2','prob1','prob2', 'probtie', 'proj_score1', 'proj_score2', 'importance1','importance2', 'score1', 'score2']]
merged_dataset.head()
```

# Regression Model

```python
X=merged_dataset[['spi1','spi2','prob1','prob2','importance1','importance2','offensive_team1','offensive_team2',"proj_score1","proj_score2"]].values
Y=merged_dataset['score1'].values
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
regressor= LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
print(y_pred)
```

```python
df=pd.DataFrame({'Actual': y_test , 'Prediction': y_pred})
df.head(10)
```

```python
print(regressor.coef_)
```

```python
er=[]
for i in range(len(y_test)):
    x=(y_test[i]-y_pred[i])**2
    er.append(x)

print('variance is:  %.2f'% np.var(er))
print('regression intercept is: %.2f'%regressor.intercept_)
print('R2 score : %.2f'% r2_score(y_test,y_pred))
print('Mean absolute error: %.2f'% metrics.mean_absolute_error(y_test,y_pred))
print('Mean squared error: %.2f'% metrics.mean_squared_error(y_test,y_pred))
print('root mean squared error: %.2f'% np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

# Random Forest

- Creating test and train data with different approach. I want to add one column to data set which is for dividing dataset to test and train.


```python
merged_dataset['for_train']=np.random.uniform(0,1,len(merged_dataset)) <= 0.8
merged_dataset.head()
```

- Now I want to create dataframes with test rows and training rows

```python

train , test = merged_dataset[merged_dataset['for_train']==True] , merged_dataset[merged_dataset['for_train']==False]
print('Number of observation in the training data: ', len(train))
print('Number of onservation in the test data: ',len(test))
```

- Now I want to select those columns which I want to use as a feature. So based on merged_dataset , columns 2 to 13 is willing to selecting.

```python
features = merged_dataset.columns[2:13]
features         # These are the features which are using 
```

```python

y1_train=train['score1']
y2_train=train['score2']
y1_test=test['score1']
y2_test=test['score2']
print('Number of observation in the training data: ', len(y1_train))
print('Number of onservation in the test data: ',len(y2_train))
print('Number of observation in the training data: ', len(y1_test))
print('Number of onservation in the test data: ',len(y2_test))

train[features]
```

```python
clf=RandomForestClassifier( n_jobs=2 , random_state=0)
clf.fit(train[features],y1_train)
clf.fit(train[features],y2_train)
```

```python
clf.predict(test[features])
```

```python
print(clf.score(train[features], y1_train))
print(clf.score(train[features], y2_train))
```

```python
print(clf.score(test[features],y1_test))
print(clf.score(test[features], y2_test))
```
